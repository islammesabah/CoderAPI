```python
from datasets import load_from_disk
```

```python
def get_data(file_path):
    test_data = load_from_disk(file_path)
    print("File Path: "+file_path)
    print(f'  ==> Loaded {len(test_data)} samples')
    return test_data
```

### Check Training data with decoder input 

50256 == End Of Sentence token

```python
data_path = "./cache_data/"
data_name = 'instructions_general_prompt'
dataset = get_data(data_path+data_name)
```

```python
#Show content of the dataset
dataset
```

```python
# Show a dataset example
input_ids = dataset["input_ids"][0]
print("input ids: ",input_ids)
print(", With length = ", len(input_ids))
attention_mask = dataset["attention_mask"][0]
print("attention mask: ",attention_mask)
print(", With length = ", len(attention_mask))
decoder_input_ids = dataset["decoder_input_ids"][0]
print("decoder input ids: ",decoder_input_ids)
print(", With length = ", len(decoder_input_ids))
decoder_attention_mask = dataset["decoder_attention_mask"][0]
print("decoder attention mask ids: ",decoder_attention_mask)
print(", With length = ", len(decoder_attention_mask))
labels = dataset["labels"][0]
print("labels: ",labels)
print(", With length = ", len(labels))
```
